### Q. 4: Let's break the data down by quarters. Create a bar graph showing the average closing stock price by quarter (quarters of all data to date, don't worry about actual financial quarters):

```python
quarter_amount = round(len(apple)/4)
first_quarter = apple.iloc[0:quarter_amount]
second_quarter = apple.iloc[quarter_amount:quarter_amount*2]
third_quarter = apple.iloc[quarter_amount*2:quarter_amount*3]
fourth_quarter = apple.iloc[quarter_amount*3:]

avg_closing_prices = [first_quarter.Close.mean(),
                      second_quarter.Close.mean(),
                      third_quarter.Close.mean(),
                      fourth_quarter.Close.mean()]
# Create a list of variances, which are set at .25 above and below the score
#variance = [df['pre_score'].mean() * 0.25, df['pre_score'].mean() * 0.25, df['pre_score'].mean() * 0.25]

variance = [first_quarter.Close.mean()* 0.25,
                      second_quarter.Close.mean()* 0.25,
                      third_quarter.Close.mean()* 0.25,
                      fourth_quarter.Close.mean()* 0.25]

# Set the bar labels
bar_labels = ['first quarter', 
              'second quarter', 
              'third quarter', 
              'fourth quarter']

# Create the x position of the bars
x_pos = list(range(len(bar_labels)))

# Create the plot bars
# In x position
plt.bar(x_pos,
        # using the data from the mean_values
        avg_closing_prices, 
        # with a y-error lines set at variance
        yerr=variance, 
        # aligned in the center
        align='center',
        # with color
        color='#42c5f4',
        # alpha 0.5
        alpha=0.5)

# add a grid
plt.grid()

# set height of the y-axis
max_y = max(zip(avg_closing_prices, variance)) # returns a tuple, here: (3, 5)
plt.ylim([0, (max_y[0] + max_y[1]) * 1.1])

# set axes labels and title
plt.ylabel('mean closing price')
plt.xticks(x_pos, bar_labels)
plt.title('Mean Quarterly Closing Price; Apple 2017')

plt.show()
```

### Q. 5: We've been given access to a time machine that can take us back to any point within the past year. We can only return for 1 week (7 days). Given Apple's stock prices, tell me what day we should go back to in order to maximize our one time investment (i.e. when could we make the most amount of money by buying as much stock as possible one day and selling it within the next week?). Additionally, assume we raise 1 million dollars in investment funds before our time hop. How much stock can we purchase on our chosen day? How much will we profit given our sale?

```python
## function to get purchase date and minimum stock value:

# first version of function, improved below

# # for index, row in test.iterrows():
# #     print(test.iloc[index])
# i = 7
# apple.iloc[i-7:i]

# def get_date_and_min(iterator,days):
#     min_val = min(apple.iloc[iterator-days:iterator].Low)
#     purchase_date = apple.loc[apple['Low'] == min_val].Date
#     return min_val, purchase_date


def get_date_and_min(iterator,days):
    if iterator == 0:
        min_val = apple.iloc[iterator]['Low']
        purchase_date = apple.iloc[iterator]['Date']
    elif iterator in list(range(1,7)):
        min_val = min(apple.iloc[0:iterator].Low) # need to use where we get this to set the date, not wherever the min matches (there are a few non-unique values)
        #purchase_date = apple.iloc[0:iterator].loc[apple.iloc[0:iterator]['Low'] == min_val].Date
        purchase_date = apple.iloc[0:iterator].loc[apple['Low'] == min_val].Date
    else:
        min_val = min(apple.iloc[iterator-days:iterator].Low)
        #purchase_date = apple.iloc[iterator-days:iterator].loc[apple.iloc[iterator-days:iterator]['Low'] == min_val].Date
        purchase_date = apple.iloc[iterator-days:iterator].loc[apple['Low'] == min_val].Date
    return min_val, purchase_date
```

```python
# go through stocks and calculate best purchase date and difference b/w high/low:

max_diff = []
purchase_dates = []
for index, row in tqdm_notebook(apple.iterrows()):
    if index == 0:
        max_diff.append(apple.iloc[0]['High']-apple.iloc[0]['Low'])
        purchase_dates.append(apple.iloc[0]['Date'])
    elif index in list(range(1,7)):
        min_val, date = get_date_and_min(index,1)
        max_diff.append(apple.iloc[index].High - min_val)
        purchase_dates.append(date[date.index[0]])
    else:
        min_val, date = get_date_and_min(index,6)
        max_diff.append(apple.iloc[index].High - min_val)
        purchase_dates.append(date[date.index[0]])

# add calculated values to dataframe after converting to series:
se = pd.Series(max_diff)
apple['difference'] = se.values
se = pd.Series(purchase_dates)
apple['purchase_date'] = se.values
```

```python
# what day we should go back to in order to maximize our one time investment
# (i.e. when could we make the most amount of money by buying as much stock 
# as possible one day and selling it within the next week?). 

# Additionally, assume we raise 1 million dollars in investment funds before our time hop. 
# How much stock can we purchase on our chosen day? How much will we profit given our sale?

maximum_difference = max(apple.difference)
index_greatest_diff = apple.loc[apple['difference'] == maximum_difference].index[0]

# let's look at our time travel itinerary:

apple.iloc[index_greatest_diff-6:index_greatest_diff+1]
```

```python
# when should we go back?
time_travel_date = apple.iloc[index_greatest_diff:index_greatest_diff+1]['purchase_date'][index_greatest_diff]
sell_date = apple.iloc[index_greatest_diff:index_greatest_diff+1]['Date'][index_greatest_diff]
print('date we go back to the future:', time_travel_date)
print('date we return to the present:', sell_date)

index_buy_date = apple.loc[apple['Date'] == time_travel_date].index[0]

# how many shares can our million dollars buy us on this date?
low_price = apple.iloc[index_buy_date]['Low']
capital = 1000000
num_shares = capital/low_price
print('we can buy ', num_shares,'shares')

# how much profit will we make?

high_price = apple.iloc[index_greatest_diff]['High']
total_return = high_price*num_shares
print('total revenue:',round(total_return,2))
print('total profit:',round(total_return-capital,2))
```
